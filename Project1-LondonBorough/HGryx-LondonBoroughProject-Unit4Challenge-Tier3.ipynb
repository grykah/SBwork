#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
properties = properties.rename(columns = {'Unnamed: 0':'Borough', pd.NaT: 'ID'})
properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
properties = pd.melt(properties, id_vars = ['Borough','ID'])
cleanProp = properties.rename(columns = {0: 'Date', 'value': 'Price'})
cleanProp.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
print('table type: ' + '\n '+ str(type(cleanProp)) + '\n')
print('table data types: '+ '\n' + str(cleanProp.dtypes) + '\n')
cleanProp['Price'] = pd.to_numeric(cleanProp['Price'])
print('updated Price data type: ' + '\n' + str(cleanProp['Price'].dtypes))
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
print('all counts: ' + '\n'+ str(cleanProp.count()))
print('boroughs: ' + '\n'+ str(cleanProp['Borough'].unique()))
print('ids: ' + '\n'+ str(cleanProp['ID'].unique()))
#print(cleanProp['Date'].unique())
#print(cleanProp['Price'].unique())

#unnamed rows
print('unnamed: ')
print(cleanProp[cleanProp['Borough'] == 'Unnamed: 47'])
print(cleanProp[cleanProp['Borough'] == 'Unnamed: 37'])
print(cleanProp[cleanProp['Borough'] == 'Unnamed: 34'])

#remove unnamed using NaNs
print('removed NaNs')
print(cleanProp[cleanProp['Price'].isna()])
cleanProp_noNa = cleanProp.dropna()
print(cleanProp_noNa.count())
```

```python
df = cleanProp_noNa
print(df.dtypes)
print(df.head())
```

```python
df_group = df.groupby(['Borough','Year']).mean()
#print(df_group.loc['City of London'])
df_group.head(30)
#df_group.dtypes
#print(df_group)
#df_group['Price'].head()

#df_selectedYears = df.loc[df['Year'] == 1995].append(df.loc[df['Year'] == 2020])
df1995 = df.loc[df['Year'] == 1995]  #used lowest prices
df1995 = df1995.reset_index(drop=True)
df2019 = df.loc[df['Year'] == 2018]  #switched to 2018 upon examination of above plot since 2018 has highest values than 2019/20, ultimately did not make a difference in the borough, but price diff was more significant
df2019 = df2019.reset_index(drop=True)

#df1995.head(50)
#df2019.tail(50)

df_priceDiff = pd.concat([df1995['Price'], df2019['Price']], axis=1)
df_priceDiff = pd.concat([df1995['Borough'], df_priceDiff], axis=1)
colName = ['Borough', 'Price 1995', 'Price 2019']
df_priceDiff.columns = colName
df_priceDiff['Price Difference'] = (df_priceDiff['Price 2019'].subtract(df_priceDiff['Price 1995']))
#df_priceDiff = df2020['Price'].subtract(df1995['Price'], fill_value=0)
#df1995.dtypes

#print(df1995['Borough'])
#print(df2020['Borough'])
print(df_priceDiff)

plt.plot(df_priceDiff['Borough'], df_priceDiff['Price Difference'], 'b*')
plt.show()

maxrow = df_priceDiff.iloc[df_priceDiff['Price Difference'].idxmax()]  #dfrm.iloc[dfrm['A'].idxmax()]
maxval = max(df_priceDiff['Price Difference'])
print('the greatest increase in price was ' + str(maxval) + ' and occurred in ' + str(maxrow['Borough']))

```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#df_group.dtypes 
#df_group.loc['City of London'].Price.iloc[24] #0-1995 24-2019

def create_price_ratio(df = df_group, Borough = 'City of London'):
    Bor = str(Borough)
    d2019 = df.loc[Bor].Price.iloc[24]
    d1995 = df.loc[Bor].Price.iloc[0]
    pRatio = float(d2019/d1995)
    return pRatio

#create_price_ratio(df_group, 'City of London')
```

```python
priceRatios = {}

for place in df_priceDiff['Borough'].unique():
    priceRatios[place] = create_price_ratio(df_group, place)
#print(priceRatios)

df_priceRatios = pd.DataFrame(priceRatios.items())
#print(df_priceRatios)
maxratinfo = df_priceRatios.iloc[df_priceRatios.iloc[:,1].idxmax()]

print('the max housing price ratio was '+str(maxratinfo[1])+' and occurred in '+str(maxratinfo[0]))
```

```python
df_pR_sort = df_priceRatios.sort_values(by = 1, ascending=False )
print(df_pR_sort)
nn = 6
bour = df_pR_sort.iloc[:,0].head(nn)
rat = df_pR_sort.iloc[:,1].head(nn)
plt.bar(bour, height = rat)
plt.plot(bour, rat)
plt.show()

print(df_pR_sort.head(nn))

```

```python
phackney = df_pR_sort.iloc[0,1]
pkinsingtonChelsea = df_pR_sort.iloc[15,1]
hackney = df_pR_sort.iloc[0,0]
kinsingtonChelsea = df_pR_sort.iloc[15,0]

plt.bar(hackney, phackney)
plt.bar(kinsingtonChelsea, pkinsingtonChelsea)
plt.show()


phackney = df_priceDiff[df_priceDiff['Borough'] == 'Hackney']['Price Difference']
hackney = df_priceDiff[df_priceDiff['Borough'] == 'Hackney']['Borough']
kc = df_priceDiff[df_priceDiff['Borough'] == 'Kensington & Chelsea']['Borough']
pkc = df_priceDiff[df_priceDiff['Borough'] == 'Kensington & Chelsea']['Price Difference']

plt.plot(hackney, phackney, 'b*')
plt.plot(kc, pkc, 'r*')
plt.show()
```

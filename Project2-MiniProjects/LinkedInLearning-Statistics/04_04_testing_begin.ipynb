```python
# however, let's account for population, and assume wide variances in the pop
# simulate areas with most cholera as a proportion to population of the area
def simulate(n):
    return pd.DataFrame({'closest': np.random.choice([0,1,4,5],size=n,p=[0.65,0.15,0.10,0.10])})
```

```python
# try it once for the total number of deaths, it is similar to previous
simulate(489).closest.value_counts()
```

```python
# simulate many samples
sampling = pd.DataFrame({'counts': [simulate(489).closest.value_counts()[0] for i in range(10000)]})
```

```python
# create a hist of the samples that assume death is proportional to population
sampling.counts.hist(histtype='step')
```

```python
# compare the death per population with observed number of cholera per pump specifically at pump 0
scipy.stats.percentileofscore(sampling.counts,340)
```

# Requests Package Use

### Requesting from a webpage

```python
import requests
```

```python
response = requests.get('https://prettyprinted.com')
#response = requests.get('https://prettyprinted.com//f') #this will fail with 404 status code.
print(response.status_code) #200 status code means everything went well, data was returned.
```

```python
if response:
    print('Response OK')
else:
    print('Response Failed')
```

```python
# header information for the webpage
print(response.headers) #gives info about server, particularly content-type is of interest so we know what/how to import data
```

```python
# page html
print(response.text) #this gives us the html of the page, beautiful soup could be used to scrape this.
```

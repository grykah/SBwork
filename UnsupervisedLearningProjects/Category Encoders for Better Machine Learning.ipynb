# Category Encoders for Better Machine Learning

```python
# import packages
import numpy as np
import pandas as pd
import category_encoders as ce
```

```python
# create some data
df = pd.DataFrame({
 'color':["a", "c", "a", "a", "b", "b", "a", "c", "c"], 
 'outcome':[1, 2, 1, 1, 3, 2, 0, 0, 1]})
print(df.head())

# split data
X = df.drop('outcome', axis = 1)
y = df.drop('color', axis = 1)
```

```python
# instantiate an encoder - binary encoder used here
ce_binary = ce.BinaryEncoder(cols = ['color'])
```

```python
# fit and transform with the encoder for encoded data
ce_binary.fit_transform(X, y)
```

```python
# utlize an alternate encoder
ce_hash = ce.HashingEncoder(cols = ['color'])
ce_hash.fit_transform(X, y)
```

```python
# utlize an alternate encoder
ce_basen = ce.BaseNEncoder(cols = ['color'], base = 1)
ce_basen.fit_transform(X, y)
```

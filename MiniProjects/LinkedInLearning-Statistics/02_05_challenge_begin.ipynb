Standard imports

```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
tb = pd.read_csv('tb.csv')
tb.head()
```

```python
tb.columns
```

```python
mlt = tb.melt(['country','year'], ['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'], 'sexForage', 'cases')
mlt.head()
```

```python
mlt['sex'] = mlt['sexForage'].str.slice(0,1)
mlt['age'] = mlt['sexForage'].str.slice(1)
mlt.head()
```

```python
mlt['age'] = mlt['age'].map({'04': '0-4', '514': '5-14', '1524': '15-24', '2534': '25-34', '3544': '35-44',
                                   '4554': '45-54', '5564': '55-64', '65': '65+', 'u': np.nan})
mlt['age'].sample(30)
```

```python
df = mlt.dropna(subset=['cases'])
df.sample(15)
```

```python
df.sort_values(['country', 'year','sex','age'])
df.head()
```

```python
df = df[['country', 'year', 'sex', 'age','cases']]
df.head()
```

```python
df.to_csv('challenge.csv')
```

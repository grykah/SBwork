## Data Cleaning

### Handle the missing and NA values
Run the code below to print out the percent of each feature that is missing. You can see from the result of the code that there are 13 columns out of 27 missing data. Most of these are missing less than 10% of the data. You may need to adapt the code if your dataframe is named something other than df.


*   Consider the columns individually 
*   Identify the appropriate fill method for each column



```python
nas=pd.DataFrame(BMRdata.isnull().sum().sort_values(ascending=False)/len(BMRdata),columns = ['percent']) #calculate the percent NaNs in data by column
pos = nas['percent'] > 0 #filter for those with NaNs
nas[pos] 
```

Let's review each of these columns in the data.


*   fastEight is the number of fast eight-person ski lifts at each of the resorts. In addition, you have the counts of all the other types of chairs at each resort and the total number of chairs as well. You can simply subtract the number of fastEight lifts from the total to determine if any of the missing values can be backfilled or if they should be equal to zero. Follow the next step below and see for yourself how to review and backfill these missing values. These columns contain all the resort lift counts by type.


* trams	
* fastEight	
* fastSixes	
* fastQuads	
* quad
* triple	
* double	
* surface	
* total_chairs <- the sum of these columns for each row.




<font color='teal'> **In order to determine an appropriate fill value for the fastEigth column, you need to calculate the difference between the sum of all the chairlift columns and the `total_chairs` column.**</font>

```python
 for row in BMRdata.iterrows():  #loop through each row
        totChairs = row[1]['trams'] + row[1]['fastSixes'] + row[1]['fastQuads'] + row[1]['quad'] + row[1]['triple'] + row[1]['double'] + row[1]['surface']
        if not row[1]['total_chairs'] == totChairs: #compare totChairs to column total chairs to see if fastEights are missing, will trigger if fastEight is present too-->look at printed data to compare
            print(str(row[1]['total_chairs'] - totChairs) + ' fastEight chairs are missing on '+ str(row[0])) #print number missing fastEight chairs
            print(row[1])
            
```

As you can see from running the above cell, there are no fastEight lift values missing because the difference between the sum of all the lift columns and the 'total_chairs' column is zero for every row in the dataframe. This confirms that you can therefore replace the missing values for the fastEight rows with zeros.

<font color='teal'> **In the cell below, replace Nan's in 'fastEight' with zero.**</font>

Hint: you could use the `fillna()` method here. 

```python
df_fillEights = BMRdata #assign df new name
df_fillEights['fastEight'].fillna(0, inplace = True) #replace NaNs with 0
#df_fillEights['fastEight'].isnull().values #double check there are no more NaNs-->should all be false
```

Consider the next column with missing values, 'NightSkiing_ac'. This column represents the number of acres of night skiing availble at this resort. Not all resorts have night skiing because it requires stadium lights and is not that popular with skiers and snowboarders. Given that information, it is also safe to assume that if there are missing values in this column, it is due to there not being any night skiing acres available at that resort so you can fill missing values with zeros safely. 

```python
df_fillNightSki = df_fillEights #assign df new name
df_fillNightSki['NightSkiing_ac'].fillna(0, inplace = True) #replace NaNs with 0
#df_fillNightSki['NightSkiing_ac'].isnull().values #double check there are no more NaNs-->should all be false
```

Next, let's consider the ticket price columns 'AdultWeekday' & 'AdultWeekend'. You know that there has to be a price for these that happens to be missing so, in this case, you cannot fill the missing values with zeros. Instead, you have a few options about what value you can use to fill the missing rows. You can simply take the mean of the column in consideration or you could backfill or forwardfill from the resort in the row before or after your missing value. Backfill and forward fill would be more reasonable in a situation where the data are in a particular order where the rows near each other are similar to each other in the price column. You could also build an imputation model to use a model to predict the value based on the data values in adjacent columns as well as adjacent rows. 

<font color='teal'> **Run the cell below to replace Nan's in 'AdultWeekday' with the mean value.**</font>

```python
df_fillPrice = df_fillNightSki #assign df new name
df_fillPrice['AdultWeekday'].fillna((df_fillPrice['AdultWeekday'].mean()), inplace=True) #fill NaNs with mean
#df_fillWeekdayPrice['AdultWeekday'].isnull().values #double check there are no more NaNs-->should all be false
df_fillPrice['AdultWeekend'].fillna((df_fillPrice['AdultWeekend'].mean()), inplace=True) #fill NaNs with mean
```

<font color='teal'> **Use what you know about the remaining columns missing data and decide whether to fill with a zero or to fill with a mean.**</font>


If no value is given for the average snowfall or the amount of snow made, fill with mean since snow is necessary for skiing and these are high snow areas.

```python
df_Snow = df_fillPrice #assign df new name
df_Snow['Snow Making_ac'].fillna((df_Snow['Snow Making_ac'].mean()), inplace=True) #fill NaNs with mean
df_Snow['averageSnowfall'].fillna((df_Snow['averageSnowfall'].mean()), inplace=True) #fill NaNs with mean
```

If days open last year is unknown, take the average.  If projected days open is unknown, use days open last year for projection.

```python
df_Open = df_Snow #assign df new name
df_Open['daysOpenLastYear'].fillna((df_Open['daysOpenLastYear'].mean()), inplace=True) #fill NaNs with mean
df_Open['projectedDaysOpen'].fillna((df_Open['daysOpenLastYear']), inplace =True) #fill NaNs with predicition = 'daysOpenLastYear'
```

If runs and longesst runs data are missing, it is likely this is an all terrain park, so fill with 0

```python
df_fillRunsLongestRuns = df_Open #assign df new name
for row in df_fillRunsLongestRuns.iterrows():
    if df_fillRunsLongestRuns['Runs'].isnull().values.any() & df_fillRunsLongestRuns['LongestRun_mi'].isnull().values.any(): #only assume all terrain if both have NaN
        df_fillRunsLongestRuns['Runs'].fillna(0, inplace = True) #replace NaNs with 0
        df_fillRunsLongestRuns['LongestRun_mi'].fillna(0, inplace = True) #replace NaNs with 0
```

If terrain parks and skiable terrain data are missing, it is likely this is an all runs park, so fill with 0.  Otherwise, this info is just missing so put average.

```python
df_terrain = df_fillRunsLongestRuns #assign df new name
for row in df_terrain.iterrows():
    if df_terrain['TerrainParks'].isnull().values.any() & df_terrain['SkiableTerrain_ac'].isnull().values.any(): #only assume all terrain if both have NaN
        df_terrain['TerrainParks'].fillna(0, inplace = True) #replace NaNs with 0
        df_terrain['SkiableTerrain_ac'].fillna(0, inplace = True) #replace NaNs with 0   
    else:
        df_terrain['TerrainParks'].fillna((df_terrain['TerrainParks'].mean()), inplace=True) #fill NaNs with mean
        df_terrain['SkiableTerrain_ac'].fillna((df_terrain['SkiableTerrain_ac'].mean()), inplace=True) #fill NaNs with mean
```

If no information is available on years open, the resort probably just opened last year, so set years open to 1.

```python
df_yOpen = df_terrain #assign df new name
df_Snow['yearsOpen'].fillna(1 , inplace=True) #fill NaNs with mean
```

Check that all NaN values have been replaced.

```python
df_yOpen.info() #-->check that all non-null counts are 330
```

Switch columns that are logically integers to int (from float).

```python
df_toint = df_yOpen #assign df new name
#change type from float to int
df_toint['projectedDaysOpen'] = df_toint['projectedDaysOpen'].astype('int64')
df_toint['fastEight'] = df_toint['fastEight'].astype('int64')
df_toint['Runs'] = df_toint['Runs'].astype('int64')
df_toint['TerrainParks'] = df_toint['TerrainParks'].astype('int64')
df_toint['daysOpenLastYear'] = df_toint['daysOpenLastYear'].astype('int64')
df_toint['yearsOpen'] = df_toint['yearsOpen'].astype('int64')
df_yOpen.info() #check info on types
```

A few items stood out from looking at the data.  There are three resorst that have over 300 runs, other parks generally have 200 or less.  These parks may just be larger or have many short runs though.

Similarly, there is a park that has over 25000 acres of skiable terrain wereas other parks all have less that 10000 acres.  This is suspicious since other parks in that region have less than 7500 acres.  This is something that should probably be double checked, but it is not impossible so I will not change it here. 

Again, there was a very high value of snow made at one resort in the Sierra Nevadas, however, there is a second resort that had a high volume of snow made, so it is possible that some resorts are at a lower elevation or had more melt or perhaps were even being used more so they required more powder to be made.  This can be rationalized so will not be changed.

In years open, there is a resort that has been open 2019 years.  This is not right.  It is much more likely that the resort opened in 2019 and that value was recorded rather than the number of years so this definately needs to be fixed.  

```python
yearRow = df_cln['yearsOpen'].max()  #get the exact value
updateVal = df_cln.loc[df_cln['yearsOpen']==yearRow] #find the row that value is in

df_cln.at[updateVal.index, 'yearsOpen'] = 1 #update row 115
df_cln.iloc[updateVal.index] #show the update
```
